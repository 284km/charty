```ruby
require 'charty'

charty = Charty::Plotter.new(:rubyplot)
```

```ruby
curve = charty.curve do
  function {|x| Math.sin(x) }
  range x: 0..10, y: -1..1
  xlabel 'foo'
  ylabel 'bar'
end
curve.render("sample_images/curve_with_function_rubyplot.png")
```

```ruby
curve2 = charty.curve do
  series [0,1,2,3,4], [10,40,20,90,70], label: "label1"
  series [0,1,2,3,4], [90,80,70,60,50], label: "label2"
  series [0,1,2,3,4,5,6,7,8], [50,60,20,30,10, 90, 0, 100, 50], label: "label3"
  range x: 0..10, y: 1..100
  xlabel 'foo'
  ylabel 'bar'
end
curve2.render("sample_images/curve_rubyplot.png")
```

```ruby
bar = charty.bar do
  series [0,1,2,3,4], [10,40,20,90,70]
  series [0,1,2,3,4], [90,80,70,60,50]
  series [0,1,2,3,4,5,6,7,8], [50,60,20,30,10, 90, 0, 100, 50]
  range x: 0..10, y: 1..100
  xlabel 'foo'
  ylabel 'bar'
  title 'bar plot'
end
bar.render("sample_images/bar_rubyplot.png")
```

```ruby
boxplot = charty.boxplot do
  data [[60,70,80,70,50], [100,40,20,80,70], [30, 10]]
  range x: 0..10, y: 1..100
  xlabel 'foo'
  ylabel 'bar'
  title 'box plot'
end
boxplot.render("sample_images/boxplot_rubyplot.png")
```

```ruby
scatter = charty.scatter do
  series 0..10, (0..1).step(0.1), label: 'sample1'
  series 0..5, (0..1).step(0.2), label: 'sample2'
  series [0, 1, 2, 3, 4], [0, -0.1, -0.5, -0.5, 0.1]
  range x: 0..10, y: -1..1
  xlabel 'x label'
  ylabel 'y label'
  title 'scatter sample'
end
scatter.render("sample_images/scatter_rubyplot.png")
```

```ruby
layout = charty.layout
```

```ruby
layout << curve
layout << scatter
layout.render("sample_images/subplot_rubyplot.png")
```

```ruby
curve_list = [0.5, 0.75].map do |f|
  charty.curve(f:f) do
    function {|x| Math.sin(f*x) }
    range x: 0..10, y: -1..1
  end
end

scatter_list = [-0.5, 0.5].map do |f|
   charty.scatter(f: f) do
    series Charty::Linspace.new(0..10, 20), Charty::Linspace.new(0..f, 20)
    range x: 0..10, y: -1..1
  end
end

grid_layout = charty.layout(:grid2x2)
grid_layout << curve_list
grid_layout << scatter_list
grid_layout.render("sample_images/subplot2_rubyplot.png")
```

```ruby
errorbar = charty.errorbar do
  series [1,2,3,4], [1,4,9,16], label: 'label1'
  series [1,2,3,4], [16,9,4,1], label: 'label2'
  range x: 0..10, y: -1..20
  yerr [0.6,0.7,2.1,1.4]
  xerr [0.1,0.2,0.3,1.4]
  xlabel 'x label'  
  title 'errorbar'
end
errorbar.render("sample_images/errorbar_rubyplot.png")
```

```ruby
bubble = charty.bubble do
  series 0..10, (0..1).step(0.1), [10, 5, 20, 20, 20, 10, 5, 0, 50, 4, 40], label: 'sample1'
  series 0..5, (0..1).step(0.2), [1, 10, 20, 10, 50, 1], label: 'sample2'
  series [0, 1, 2, 3, 4], [0, -0.1, -0.5, -0.5, 0.1], [40, 30, 20, 10, 5], label: "sample3"
  range x: 0..10, y: -1..1
  xlabel 'x label'
  ylabel 'y label'
  title 'bubble sample'
end
bubble.render("sample_images/bubble_rubyplot.png")
```

```ruby
hist = charty.hist do
  data [[10, 10, 20, 30, 40, 40,40,40,40,40, 50, 10, 10, 10], [100, 100, 100, 100, 90, 90, 80, 30, 30, 30, 30, 30]]
  range x: 0..100, y: 0..7
  xlabel 'x label'
  ylabel 'y label'
  title 'histogram sample'
end
hist.render("sample_images/hist_rubyplot.png")
```
